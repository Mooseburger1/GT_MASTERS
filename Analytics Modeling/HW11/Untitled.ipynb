# Question 15.2
---------------------------
In the videos, we saw the “diet problem”. (The diet problem is one of the first large-scale optimization
problems to be studied in practice. Back in the 1930’s and 40’s, the Army wanted to meet the nutritional
requirements of its soldiers while minimizing the cost.) In this homework you get to solve a diet problem
with real data. The data is given in the file diet.xls.
1. Formulate an optimization model (a linear program) to find the cheapest diet that satisfies the
maximum and minimum daily nutrition constraints, and solve it using PuLP. Turn in your code
and the solution. (The optimal solution should be a diet of air-popped popcorn, poached eggs,
oranges, raw iceberg lettuce, raw celery, and frozen broccoli. UGH!)

### Variables

-----------------

n - foods

m - nutrients

a$_{ij}$ = amount of nutrient j per unit of food i

m$_{j}$ = minimum daily intake of nutrient j

M$_{j}$ = maximum daily intake of nutrient j

c$_{i}$ = per-unit cost of food i

x$_{i}$ = amount of food i in daily diet

## Optimization model variables
x$_{i}$ = amount of food i in daily diet




## Constraints
$\sum_{i} a_{ij} x_i$ $\geq$ m$_{j}$ for each nutrient j

$\sum_{i} a_{ij} x_i$ $\leq$ M$_{j}$ for each nutrient j

x$_{i}$ $\geq$ 0 for each food i



## Objective function
Minimize $\sum_{i} c_ix_i$

 -------------------------------------------------

#### Let's first instantiate the optimization model and create the values {**n, m, i, j, c$_i$, a$_i$$_j$, x$_i$, m$_j$, M$_j$**}

```python
#Instantiate pulp model
opt_model = plp.LpProblem(name = 'Army Diet')
```

```python
# Number of foods
n = len(df)

# Number of nutrients
m = len(df.columns)

#helpful iterators
set_I = range(1, n+1)
set_J = range(1, m+1)

#cost for food per serving
c = {i:list(data['Price/ Serving'])[i-1] for i in set_I}

#value of each nutrient for each food
a = {(i,j): df.values[i-1][j-1] for i in set_I for j in set_J}

#minimum thresholds for each nutrient
min_ = {j:list(constraints.iloc[0])[j-1] for j in set_J}

#maximum thresholds for each nutrient
Max = {j:list(constraints.iloc[1])[j-1] for j in set_J}
```

```python
# x variables - dictionary which containes pulp objects
x_vars = {i:plp.LpVariable(cat=plp.LpContinuous,
                           lowBound = 0,
                           name=data['Foods'][i-1]) for i in set_I}
```

Next, we need to create the minimum and maximum nutrient thresholds, and add them to the model

```python
min_constraints = {j:plp.LpConstraint(
                    e = plp.lpSum(a[i,j] * x_vars[i] for i in set_I),
                    sense = plp.LpConstraintGE,
                    rhs = min_[j],
                    name = 'min_{}'.format(constraints.columns[j-1])
                    )
                  for j in set_J
                 }
```

```python
for j in set_J:
    opt_model += min_constraints[j]
```

```python
max_constraints = {j:plp.LpConstraint(
                    e = plp.lpSum(a[i,j] * x_vars[i] for i in set_I),
                    sense = plp.LpConstraintLE,
                    rhs = Max[j],
                    name = 'max_{}'.format(constraints.columns[j-1])
                    )
                   for j in set_J
                  }
```

```python
for j in set_J:
    opt_model += max_constraints[j]
```

Next, we create the objective function and add it to the model

```python
objective = plp.lpSum(x_vars[i] * c[i] for i in set_I)
```

```python
opt_model.sense = plp.LpMinimize
opt_model.setObjective(objective)
```

# Model Summary

```python
opt_model
```

# Solve the model and display output

```python
opt_model.solve()
```

```python
opt_df = pd.DataFrame.from_dict(
                                x_vars, 
                                orient='index', 
                                columns=['variable_object']
                               )

opt_df['solution_value'] = opt_df['variable_object'] \
                           .apply(lambda item: item.varValue)
```

```python
opt_df = opt_df[opt_df.solution_value != 0]
opt_df['Price'] = [c[p] for p in opt_df.index]
opt_df['Cost'] = opt_df.solution_value * opt_df.Price
```
