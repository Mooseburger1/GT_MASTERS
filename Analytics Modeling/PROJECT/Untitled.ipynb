```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
non = np.random.randint(low = 3, high = 6, size = (13))
```

```python
imp = [5,3,4,5,6,7,5,200,6,3,100,8,5]
```

```python
plt.figure(figsize = (15,8))
plt.bar(range(len(non)),non)
plt.xlabel('Piece of Clock Being Drawn')
plt.ylabel('delta-T (microseconds)')
plt.title('Delta-T Measurements for Non Impaired Patient')
plt.ylim(ymax = 50)
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.bar(range(len(imp)), imp)
plt.xlabel('Piece of Clock Being Drawn')
plt.ylabel('delta-T (microseconds)')
plt.title('Delta-T Measurements for Impaired Patient')
plt.arrow(x = 4, y = 175, dx = 2, dy = 0, head_width = 5, head_length = 0.5, color = 'k')
plt.arrow(x = 4, y = 75, dx = 5, dy = 0, head_width = 5, head_length = 0.5, color = 'k')
plt.annotate(xy = (0.3,174), s = 'Outlier in delta-T (potential mental lapse)')
plt.annotate(xy = (0.3,74), s = 'Outlier in delta-T (potential mental lapse)')
plt.show()
```
